```python
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/ds4.csv',encoding="ISO-8859-1")
```

```python
df.head()
```

```python
!pip install pgmpy
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.models import BayesianModel
from pgmpy.inference import VariableElimination
```

```python
print(df)

model = BayesianModel([
    ('age', 'Lifestyle'),
    ('Gender', 'Lifestyle'),
    ('Family', 'heartdisease'),
    ('diet', 'cholestrol'),
    ('Lifestyle', 'diet'),
    ('cholestrol', 'heartdisease'),
    ('diet', 'cholestrol')
])
```

```python
model.fit(df, estimator=MaximumLikelihoodEstimator)
df_infer=VariableElimination(model)
```

```python
print('For Age enter SuperSeniorCitizen:0, SeniorCitizen:1, MiddleAged:2, Youth:3, Teen:4')
print('For Gender enter Male:0, Female:1')
print('For Family History enter Yes:1, No:0')
print('For Diet enter High:0, Medium:1')
print('for LifeStyle enter Athlete:0, Active:1, Moderate:2, Sedentary:3')
print('for Cholesterol enter High:0, BorderLine:1, Normal:2')
```

```python
q=df_infer.query(variables=['heartdisease'], evidence={
    'age': int(input('Enter Age: ')),
    'Gender': int(input('Enter Gender: ')),
    'Family': int(input('Enter Family History: ')),
    'diet': int(input('Enter Diet: ')),
    'Lifestyle': int(input('Enter Lifestyle: ')),
    'cholestrol': int(input('Enter Cholestrol: '))
})

print(q)
```

### Diabetic dataset

```python
dia=pd.read_csv('/content/drive/MyDrive/diabetes (1).csv',encoding="ISO-8859-1")
```

```python
dia.head()
```

```python
import numpy as np
print(dia)
```

```python
dia['Glucose'] = dia['Glucose'].apply(lambda x: 0 if x < 50 else (1 if x < 100 else (2 if x < 150 else 3)))
dia.head()
```

```python
print(min(dia['BloodPressure']),max(dia['BloodPressure']))
```

```python
dia['BloodPressure'] = dia['BloodPressure'].apply(lambda x: 0 if x < 40 else (1 if x < 80 else 2))
dia.head()
```

```python
print(min(dia['BMI']),max(dia['BMI']))
```

```python
dia['BMI'] = dia['BMI'].apply(lambda x: 0 if x < 15.0 else (1 if x < 30.0 else (2 if x<45.0 else (3 if x<60.0 else 4))))
print(dia)
```

```python
print(min(dia['Age']),max(dia['Age']))
```

```python
dia['Age'] = dia['Age'].apply(lambda x: 0 if x >=21 and x<35 else (1 if x >=35 and x<60 else (2)))
print(dia)
```

```python
print(min(dia['Pregnancies']),max(dia['Pregnancies']))
```

```python
dia['Pregnancies'] = dia['Pregnancies'].apply(lambda x: 0 if x<3 else (1 if x<6 else (2)))
print(dia)
```

```python
model1= BayesianModel([
    ('Pregnancies', 'Glucose'),
    ('Glucose', 'BloodPressure'),
    ('Glucose', 'BMI'),
    ('BloodPressure', 'Outcome'),
    ('BMI', 'Outcome'),
    ('Age', 'Outcome')
])
```

```python
model1.fit(dia, estimator=MaximumLikelihoodEstimator)
diabetes_infer = VariableElimination(model1)
```

```python
q=diabetes_infer.query(variables=['Outcome'], evidence={
    'Pregnancies': int(input('Enter Pregnancies (0:): ')),
    'Glucose': int(input('Enter Glucose: ')),
    'BloodPressure': int(input('Enter Blood Pressure: ')),
    'BMI': float(input('Enter BMI: ')),
    'Age': int(input('Enter Age: ')),
})

print(q)
```
